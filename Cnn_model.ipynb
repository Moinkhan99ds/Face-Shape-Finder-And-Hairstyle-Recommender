```python
import os

os.makedirs("face_shape_dataset", exist_ok=True)
```

```python
face_shapes = ['oval', 'round','square','heart','diamond']

for shape in face_shapes:
    os.makedirs(f"face_shape_dataset/{shape}", exist_ok=True)

print(" All folders created successfully!")
```

```python

import tensorflow as tf
from tensorflow.keras import layers, models

img_size = 128

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "face_shape_dataset",       #folder ka naam
    image_size=(img_size, img_size),
    batch_size=32,
    label_mode="categorical"
)

cnn_model = models.Sequential([

    layers.Conv2D(32, (3,3), activation='relu', input_shape=(img_size, img_size, 3)),
    layers.MaxPooling2D(2,2),


    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),


    layers.Flatten(),


    layers.Dense(128, activation='relu'),


    layers.Dense(5, activation='softmax')
])


cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = cnn_model.fit(train_ds, epochs=10)


cnn_model.save("face_shape_cnn_model.h5")

print(" Model successfully trained and saved as 'face_shape_cnn_model.h5'")
```

```python

from tensorflow.keras.models import load_model


model = load_model("face_shape_cnn_model.h5")

loss, acc = model.evaluate(train_ds)
print("Loaded model accuracy:", acc * 100, "%")
```
